```python
import kagglehub
import shutil
import os

# Download the dataset
path = kagglehub.dataset_download("mewbius/ecommerce-products")

# Specify the destination path
destination_path = "/content/dataset"

# Create the destination directory if it doesn't exist
os.makedirs(destination_path, exist_ok=True)

# Move the downloaded files to the destination path
for file_name in os.listdir(path):
    shutil.move(os.path.join(path, file_name), destination_path)

print("Path to dataset files:", destination_path)
```

```python
data['combined'] = data['name'] + " " + data['description']
```

```python
new_data=data[["Classification","combined"]]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
vector1=cv.fit_transform(new_data["combined"].values.astype('U')).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector1)
```

```python
similarity
```

```python
new_data['url'] = data["url"]
```

```python
new_data
```

```python
new_data[new_data['Classification'] == "Pants"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].Classification)
```

```python
def recommand(products):
    # Get the index of the product
    index = new_data[new_data['Classification'] == products].index[0]

    # Compute similarity distances
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])

    # Display the top 5 recommendations with their links
    for i in distance[1:6]:  # Skip the first one as it will be the same product
        classification = new_data.iloc[i[0]]['Classification']
        link = new_data.iloc[i[0]]['url']
        print(f"Product: {classification}, Link: {link}")
```

```python
recommand("Blankets")
```
